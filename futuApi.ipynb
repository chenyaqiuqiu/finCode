```python
from futu import *

Host='127.0.0.1'
Port=10030
# TrdMarket.HK
# TrdSide.BUY

def connect_OpenD(market):
    return OpenSecTradeContext(filter_trdmarket=market, host=Host, port=Port, security_firm=SecurityFirm.FUTUSECURITIES)

def close_OpenD(trd_ctx):
    trd_ctx.close()  # 关闭当条连接

def acc_code_list(trd_conn, env):
    ret, data = trd_conn.position_list_query(trd_env=env)
    if ret == RET_OK:
        #print(data)
        if data.shape[0] > 0:  # 如果持仓列表不为空
            #print(data['code'])  # 获取持仓第一个股票名称
            scode = data['code'].str.split('.').str.get(1)
            list = [str(obj) for obj in scode]
            print(list)
    else:
        print('position_list_query error: ', data)

    return list

# 模拟仓： 获取所有未完成订单
def get_unfinish_simulate_orders(trd_ctx):
    ret, data = trd_ctx.order_list_query(trd_env=TrdEnv.SIMULATE)
    if ret == RET_OK:
        print(data)
    else:
        print('order_list_query error: ', data)
    return data
    
# 模拟仓： 获取所有未完成订单列表
def get_unfinish_simulate_orders_list(trd_ctx):
    ret, data = trd_ctx.order_list_query(trd_env=TrdEnv.SIMULATE)
    if ret == RET_OK:
        print(data)
        if data.shape[0] > 0:  # 如果订单列表不为空
            print(data['order_id'][0])  # 获取未完成订单的第一个订单号
            list = data['order_id'].values.tolist()
            print(list)  # 转为 list
    else:
        print('order_list_query error: ', data)
    return list

# 模拟仓： 获取处于提交状态的订单
def get_submitted_simulate_orders_list(trd_ctx):
    ret, data = trd_ctx.order_list_query(trd_env=TrdEnv.SIMULATE)
    if ret == RET_OK:
        if data.shape[0] > 0:  # 如果订单列表不为空
            submitted_orders = data[data['order_status'] == 'SUBMITTED']
    else:
        print('order_list_query error: ', data)
    return submitted_orders

# 模拟仓： 取消所有订单，目前不支持模拟仓
# cancel all simulate orders is not support
def cancel_all_simulate_orders(trd_ctx):
    ret, data = trd_ctx.cancel_all_order(trd_env=TrdEnv.SIMULATE)
    if ret == RET_OK:
        print(data)
    else:
        print('cancel_all_order error: ', data)

# 模拟仓： 取消所有订单，按照订单编号列表
def cancel_simulate_orders(trd_ctx, orders):
    for order_id in orders:
        ret, data = trd_ctx.modify_order(ModifyOrderOp.CANCEL, order_id, 0, 0, trd_env=TrdEnv.SIMULATE)
        if ret == RET_OK:
            print('order_id ', order_id, ' is cancelled')
        else:
            print('modify_order ',order_id, ' error: ', data)

###  -----  下单， 取消订单 ---- ###
# 模拟仓 下单API
def put_simulate_order(trd_ctx, price, num, code, bs):
    ret, data = trd_ctx.place_order(price=price, qty=num, code=code, trd_side=bs, trd_env=TrdEnv.SIMULATE)
    if ret == RET_OK:
        print(data)
        #print(data['order_id'][0])  # 获取下单的订单号
        #print(data['order_id'].values.tolist())  # 转为 list
    else:
        print('place_order error: ', data)
        
# 模拟仓： 取消在提交中的订单
def cancel_simulate_subbmitted_orders(trd_ctx):
    data = get_submitted_simulate_orders_list(trd_ctx)
    list = data['order_id'].values.tolist()
    cancel_simulate_orders(trd_ctx, list)
    
def apiTest():
    trd_ctx = connect_OpenD(TrdMarket.HK)
    #tickers = acc_code_list(trd_ctx, TrdEnv.SIMULATE)
    #put_simulate_order(trd_ctx, 200, 100, "HK.00700", TrdSide.BUY)
    #un_list = get_unfinish_simulate_orders(trd_ctx)
    #cannel_all_simulate_orders(trd_ctx)
    #cancel_simulate_orders(trd_ctx, un_list)
    #un_list = get_submitted_simulate_orders_list(trd_ctx)
    cancel_simulate_subbmitted_orders(trd_ctx)
    close_OpenD(trd_ctx)

apiTest()
```
